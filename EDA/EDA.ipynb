```python
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.svm import SVR
```

```python
x = df_model.drop(columns=['SalePrice'])
y = df_model['SalePrice']
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_model.drop(columns=['SalePrice']), df_model['SalePrice'], test_size=0.2, random_state=42,shuffle=True)
```

```python
xgb_model = XGBRegressor(
    n_estimators=100,       # jumlah pohon
    learning_rate=0.1,      # step shrinkage
    max_depth=5,            # kedalaman pohon
    random_state=42
)

xgb_model.fit(x_train, y_train)

# Evaluasi
y_pred_xgb = xgb_model.predict(x_test)
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Definisi fungsi MAPE
def mean_absolute_percentage_error(y_true, y_pred):    
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    # Hindari pembagian nol
    non_zero_idx = y_true != 0
    return np.mean(np.abs((y_true[non_zero_idx] - y_pred[non_zero_idx]) / y_true[non_zero_idx])) * 100
```

```python
print("XGBoost:")
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_xgb)))
print("MAE :", mean_absolute_error(y_test, y_pred_xgb))
print("R²  :", r2_score(y_test, y_pred_xgb))
print("MAPE:", mean_absolute_percentage_error(y_test, y_pred_xgb))
```
